```python
print(result[0])
```

```python
nodes = []
nodes_dict ={}
for r in result:
    for k ,v in r.items():
        if type(v) == dict: #is Node
            node_id = v['documentId']
            if node_id not in nodes_dict:
                nodes_dict[node_id] = {}
                nodes_dict[node_id]['id'] = node_id
                nodes_dict[node_id]['name'] = v['text']
                nodes_dict[node_id]['entityType'] = v['NodeType']
                nodes_dict[node_id]['externalId'] = v['id']
                nodes.append(nodes_dict[node_id])
        elif type(v) == tuple: # is link
            node_id = v[0]['documentId']
            node = nodes_dict[node_id]
            node[v[1]] = v[2]['text']
print(nodes)
```
